In traditional neurons, weights must be set manually, later on, this problem has been adressed by creation of improved neurons - that is PERCEPTRONS. if weeights of neurons are higher than threshold then neuron activates namely we call these functions as activation functions

forward pass - pass from the input of perceptron to output. this is done by STOCHASTIC GRADIENT DECENT

how the train stage works - we give the perceptron our input, we get output then we compare to label. if they are not same,difference will be computed and propogate back to neural netword. it is called BACKWARD PROPOGATION

basically, first sample forward pass, we compare the labels and compute the error, backward pass, we propogate the eror through the network
then second sample, forward pass with updated weight ... aim is to get highest accuracy
