```python
"""
LSTM - we use it for analizing the sequencial data. For example, translator.
RNN does not work well with long sequence. since it is recursive network, if you have long recursive loop,
it may cause vanishing gradient, exploding gradient. in this case, model does not learn.
solution: LSTM (long short term memory)
  input gate
  forget gate
  output gate
GRU has similar performance as LSTM
"""
```

```python
import torch
from torch import nn
```

```python
""""
Input:
input,h_0,c_0

input = the data
h_0 = initial hidden state of the elements in the input data
c_0 = initial state for the cells

input = (L,N,H_in)
L - sequence length
N - batch size
H_in - input size of data

#######

h_0 = (D, H_out)
c_0= = (D,num_cells)

D - number of directions
H_out = number of features in output
"""
```

```python
#create single lstm node

lstm_node = nn.LSTM(10,20,1) #numer of rows of our data, output, number of lsdm cell

#create the input

input_data = torch.randn(5,3,10)

h_0 = torch.randn(1,3,20)
c_0 = torch.randn(1,3,20)

output,(h_n,c_n) = lstm_node(input_data,(h_0,c_0))

print(output.shape)

```
