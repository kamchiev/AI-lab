Support Vector Machines - mostly used in supervised learning, discriminative classification algorithm. if the cloud of points are not linearly seperable, we may use radial basis function to map points into higher dimention,and then do the splitting. it is called kernel trick

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
import cv2
from sklearn.cluster import KMeans
```

```python
#load the image
# we map image color to lower space
img_arr = cv2.imread("")
(h,w,c) = img_arr.shape
```

```python
#convert img 3d -> 3d
img2D = img_arr.reshape(h*w,c)
kmeans_model = KMeans(n_clusters=16) # 16 is number of colors we want

cluster_labels = kmeans_model.fit_predict(img2D)
# convert points coordinates to good pixel values
rgb_cols = kmeans_model.cluster_centers_.round(0).astype(int)
# get back 3d shape image
img_quant = np.reshape(rgb_cols[cluster_labels],(h,w,c))
cv2.imshow()
cv2.waitKey(0)
```
