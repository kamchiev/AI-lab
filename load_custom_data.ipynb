```python
import os
import pandas as pd
from torchvision.io import read_image
from torch.utils.data import Dataset
```

```python
class MyDatasetLoader(Dataset):
  def __init__(self,labels,img_dir,transform = None):

    self.img_dir = img_dir
    self.transform = transform
    self.labels = pd.read_csv(labels)

  def __len__(self):
    return len(self.labels) #returns number of labels

  def __getitem__(self,index):
    #create the images path
    img_path = os.join(self.img_dir,self.labels.iloc[index,0])
    #read the image through its path
    image = read_image(img_path)
    #read the label
    label = self.labels.iloc[index,1]

    if self.transform:
      image = self.transform(image)

    return image, label
```
