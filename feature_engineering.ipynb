```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
import pandas as pd
```

```python
data = [
    {"price":12345,"room":4,"neighbourhood":"Via roma"},
    {"price":84656,"room":6,"neighbourhood":"Via fabiani"},
    {"price":44444,"room":4,"neighbourhood":"Appia"}
]
"""
{"Via roma":1,"Via fabiani":2,"Appia":3} # we can do that bc our model thinks one is higher than other. we may use one hot encoding
"""
```

```python
vec = DictVectorizer(dtype=int)
new_features= vec.fit_transform(data)

print(vec.get_feature_names_out())
print(new_features)
#
```

```python
data = [
    "problem of evil",
    "evil queeen",
    "horizon problem"
]

vec = CountVectorizer()
X = vec.fit_transform(data)
#covert the vectorized data to pandas dataframe
vis_data = pd.DataFrame(X.toarray(),columns=vec.get_feature_names_out)

print(X)
```

```python
# TF-IDF = term frequnecy - inverse document frequency
vec2 = TfidfVectorizer()
X2 = vec2.fit_transform(data)
vis_data = pd.DataFrame(X2.toarray(),columns = vec2.get_feature_names_out())
print(vis_data)
```
